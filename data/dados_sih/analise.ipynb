```python
df_sem_tratamento = sidrapy.get_table(
    table_code='5938',
    territorial_level='3',
    variable='37',
    ibge_territorial_code='all',
    period='all'
    
    
)
df_sem_tratamento.head()
```

```python
df_sem_tratamento.columns = df_sem_tratamento.iloc[0]
df_sem_tratamento.head()
```

```python
df_sem_tratamento = df_sem_tratamento.iloc[1:]
df_sem_tratamento.head()
```

```python
população_residente = df_sem_tratamento.pivot_table(
    index='Unidade da Federação',
    columns='Variável',
    values='Valor',
    aggfunc='sum'
).reset_index()
população_residente.head()
```

soma total do PIB por ano 

```python
soma = df_sem_tratamento.groupby('Ano')['Valor'].sum()
display(soma)
```

soma total do PIB por estados 

```python
soma_estado = df_sem_tratamento.groupby('Unidade da Federação')['Valor'].sum().sort_values(ascending=False)
display(soma_estado)
```

```python
# Converte para número — trata casos com vírgula e ponto
df_sem_tratamento['Valor'] = (
    df_sem_tratamento['Valor']
    .astype(str)
    .str.replace('.', '', regex=False)     
    .str.replace(',', '.', regex=False)   
    .astype(float)                         
)
```

```python
pd.options.display.float_format = '{:,.0f}'.format
```

 PIB médio por estado

```python
PIB_medio = df_sem_tratamento.groupby('Unidade da Federação')['Valor'].mean()
display(PIB_medio)
```

PIB por estado e ano

```python
df_pivot = df_sem_tratamento.pivot_table(values='Valor', index='Unidade da Federação', columns='Ano', aggfunc='sum')
display(df_pivot)
```
